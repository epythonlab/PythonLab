```python
import pandas as pd
```

```python
df = pd.read_csv('CustomerMaster_x08_V1_20220602081849.csv', error_bad_lines=False, engine='python', sep='delimiter')
```

```python
df.head()
```

```python
# split and separate the columns
def split_column(df, column_name):
    df['CUSTOMERNUMBER'] = df[column_name].apply(lambda x: x.split(',')[0])
    df['CUSTOMERNAME'] = df[column_name].apply(lambda x: x.split(',')[1])
    df['DELETIONINDICATOR'] = df[column_name].apply(lambda x: x.split(',')[2])
    # drop the old column
    df = df.drop(column_name, axis=1)

    return df
        
```

```python
df = split_column(df, 'CUSTOMERNUMBER,CUSTOMERNAME,DELETIONINDICATOR')
```

```python
df.head()
```

```python
# parse the data and strip quotations from the data values
import re
def strip_quotes(columns, df):
    for col in columns:
        df[col] = df[col].apply(lambda x: re.sub(r'[^\w\s]', '', x))
        
    return df
```

```python
df = strip_quotes(['CUSTOMERNUMBER', 'CUSTOMERNAME', 'DELETIONINDICATOR'], df)
```
