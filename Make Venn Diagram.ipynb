```python
# import the module
from matplotlib_venn import venn2, venn3
from matplotlib import pyplot as plt
```

```python
# define a function to make a venn diagrams
def make_venn(venn, lst, labels):
    
    # make venn and compare different lists
    return venn(lst, set_labels=labels)
```

```python
# make lists and compare
genes_list1 = ["gene1", "gene2", "gene3", 
               "gene4", "gene5", "gene6",
               "gene7", "gene8", "gene9", "gene10"]

genes_list2 = ["gene0", "gene1", "gene2", "gene3", 
               "gene8", "gene9", "gene10", "gene11", 
               "gene12", "gene13"]
```

```python
# compare the two lists by venn diagrams
make_venn(venn2, [set(genes_list1), set(genes_list2)],
         ('Drug A Response', 'Drug B Response'))
plt.title('Comparison of Drug responsive genes')
plt.show()
```

```python
genes_list3 = ["gene0", "gene4", "gene5", 
               "gene8", "gene9", "gene10",
               "gene11", "gene13", "gene14",
               "gene15", "gene16"]

```

```python
# compare the two lists by venn diagrams
labels = ['Drug A Response', 'Drug B Response', 'Drug C Response']
make_venn(venn3, [set(genes_list1), set(genes_list2), 
                  set(genes_list3)],
         set(labels))
plt.title('Comparison of Drug responsive genes')
plt.show()
```

```python
genes_list4 = ["gene1", "gene4", "gene5", 
               "gene14", "gene15", "gene17",
               "gene18", "gene19"]
```

```python
# compare the two lists by venn diagrams
labels = ['Drug A Responsive', 'Drug B Responsive', 'Drug D Responsive']
make_venn(venn3, [set(genes_list1), set(genes_list2), 
                  set(genes_list4)],
         set(labels))
plt.title('Comparison of Drug responsive genes')

plt.show()
```
